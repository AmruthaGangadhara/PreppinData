#### Split the Transaction Code to extract the letters at the start of the transaction code. These identify the bank who processes the transaction

```python
#Used lambda expression to split each string by '-' and slicing the first part in the result.
df['Bank']=df['Transaction_Code'].apply(lambda x: x.split('-')[0])
df
```

#### Rename the values in the Online or In-person field, Online of the 1 values and In-Person for the 2 values. 

```python
#using the replace function within pandas
df['Online_or_In_Person']=df['Online_or_In_Person'].replace([1,2],['Online','In Person'])
df
```

#### Change the date to be the day of the week

```python
#converting the date to day name using strftime function
df['day_name'] = pd.to_datetime(df['Transaction_Date']).dt.strftime('%A')
```

```python
df
```

#### Different levels of detail are required in the outputs. You will need to sum up the values of the transactions in three ways (help):
1. Total Values of Transactions by each bank
2. Total Values by Bank, Day of the Week and Type of Transaction (Online or In-Person)
3. Total Values by Bank and Customer Code

#### 1.Total Values of Transactions by each bank

```python
values_by_bank= df.groupby('Bank')['Value'].sum().reset_index(name='Values')

values_by_bank.head()
```

#### 2.Total Values by Bank, Day of the Week and Type of Transaction (Online or In-Person)

```python
values= df.groupby(['Bank','Online_or_In_Person','day_name'])['Value'].sum().reset_index(name='Values')
values
```

#### 3.Total Values by Bank and Customer Code

```python
bank_cust=df.groupby(['Bank','Customer_Code'])['Value'].sum().reset_index(name='Total_values')
bank_cust
```
